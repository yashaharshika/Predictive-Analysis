10) Is there any relation between Age and BMI?

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming 'df_cleaned' is your DataFrame containing the dataset
# Selecting 'Age' and 'BMI' columns
X = df_cleaned[['Age', 'BMI']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow method to determine the optimal number of clusters
inertia = []
k_range = range(1, 11)  # Testing k from 1 to 10 clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the elbow plot
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o', linestyle='-', color='b')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Plot for Optimal k')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X_scaled)

# Add cluster labels to the DataFrame
df_cleaned['Cluster'] = kmeans.labels_

# Plotting the clusters
plt.figure(figsize=(8, 6))
for cluster in range(optimal_k):
    cluster_data = df_cleaned[df_cleaned['Cluster'] == cluster]
    plt.scatter(cluster_data['Age'], cluster_data['BMI'], label=f'Cluster {cluster + 1}')

# Plotting the centroids
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('K-means Clustering of Age vs BMI')
plt.legend()
plt.grid(True)
plt.show()
```

4) Will performing Principal Component Analysis (PCA) speed up the prediction process?

```python
from sklearn.decomposition import PCA

# splitting features and target variables
X = df_cleaned.drop(columns=['Brain Stroke'])
y = df_cleaned['Brain Stroke']

# Encode categorical features
X_encoded = pd.get_dummies(X)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)
```

```python
# performing PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# variance plot
plt.figure(figsize=(6, 4))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Variance Explained by Principal Components')
plt.grid(True)
plt.show()
```

```python
# Print loadings (PCA components)
loadings = pca.components_
print("PCA Loadings:")
print(pd.DataFrame(loadings, columns=X_encoded.columns))

# Print score values (transformed data)
scores = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
print("PCA Score Values:")
print(scores)
```
